Renombramos las columnas

```
df.columns = ['Indice','Tipo_Operacion', 'Tipo_Propiedad', 'Localidad','Ubicacion_Completa','Pais','Provincia','Geonames','Latitud-Longitud','Latitud','Longitud','Precio','Moneda','Precio_Calc_ARS','Precio_Calc_USD','Sup_total_m2','Sup_Cubierta_m2','USD_por_m2_total','USD_por_m2_cub','Piso','Habitaciones','Expensas','Prop_URL','Descripcion','Titulo','Imagen']
```

En la columna 'Tipo_Propiedad' traducimos al espanol.

```
df.loc[df['Tipo_Propiedad'] == 'apartment', 'Tipo_Propiedad'] = 'Departamento'
df.loc[df['Tipo_Propiedad'] == 'PH', 'Tipo_Propiedad'] = 'Ultimo_Piso'
df.loc[df['Tipo_Propiedad'] == 'house', 'Tipo_Propiedad'] = 'Casa'
```

(OPCIONAL)filtramos los que **NO** tienen precio en USD por M2  - Cuando tengamos las REGEX vamos a poder hacer este paso despues de "salvar" registros usando la descripcion.

```
df2 = df2.loc[df2.USD_por_m2_total.notnull(),:]
```

```
(count, division)= np.histogram(df2.USD_por_m2_total)
bins = []
for i in range(0,len(division)-1):
    bins.append((division[i],division[i+1]))

bins = pd.DataFrame({'bin': bins,
                     'cantidad': count})
bins
```
